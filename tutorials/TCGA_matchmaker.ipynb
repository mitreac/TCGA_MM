# Transcriptome analysis of cancer samples from the The Cancer Genome Atlas 
## Match of the cancer samples to a reference transcriptome profile

### Import the library 

```python
from TCGA_matchmaker import match_computation as m
import pandas as pd
```

Example of distance computation

The reference transcriptomic profile

```python
profile = pd.Series([2,3,4], index = ["g1","g2","g3"])
profile
```

The cancer sample transcriptomic profile

```python
sample_data = pd.Series([2,4,4,5], index = ["g1","g2","g4","g5"])
sample_data
```

Computing the distance between the cancer sample and reference profile

```python
distance = m.compute_distance(profile, sample_data)
distance
```
